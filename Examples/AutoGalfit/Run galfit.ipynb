```python
from PyRite.AutoGalfit import *
```

```python
ag = AutoGalfitClass("f277w.fits",
                     26.51323677438482, ext = 0,
                     imageRegion=[150, 250, 150, 250])
ag.psfFile = "PSF_F277W.fits"
ag.prepareAuxFiles()
ag.prepareConsFile()
ag.addCon("2-1 x -2 2")
ag.addCon("2-1 y -2 2")
```

```python
ag.showImage(maskOn=1, scale = 'log')
```

```python
ser = FitObject("sersic",
                mag = 17, randomMag=0.2,
                apix = 5, randomApix=1,
                n = 3, randomN=1,
                ba = 0.7, randomBA=0.2,
                randomPA = 45, randomX = 5, randomY = 5)
psf = FitObject('psf',
                mag=19, randomMag=0.2,
                randomX=3, randomY=5
                )
ag.fitObjects = [ser, psf]
ag.galfitBootstrap(n = 10, saveModels = True)
```
