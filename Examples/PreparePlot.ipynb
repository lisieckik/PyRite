```python
from PyRite.plotting import set_style, updateAxisColor, remakeTicks, makeSubplots
import matplotlib.pyplot as plt
```

```python
set_style()
fig = plt.figure()
axes = makeSubplots(fig, 2,2,[0.1, 0.07,0.01,0.01],
                    procentagey = [70, 30], procentagex = [70,30],
                    visibility=[1,1,1,0], xspace = 0.02, yspace = 0.03,
                    colorbar=['y',0.03,0.1])
```

```python
set_style(style='dark')
fig = plt.figure()
axes = makeSubplots(fig, 2,2,[0.1, 0.07,0.01,0.01],
                    procentagey = [70, 30], procentagex = [70,30],
                    visibility=[1,1,1,0], xspace = 0.02, yspace = 0.03)
remakeTicks(0.3, 0.25, ax = axes[0], yoff = 0.1)
```
